```python
data = pd.read_csv('../input/xrpxlmxdctel/xrp-xlm-tel-xdc-04_15_18-10_03_21.csv')

correlationData = data[['xrp_price','xlm_price','xdc_price','tel_price']]

def linearRegression(symbol1,symbol2):
    
   df2 = pd.DataFrame({'tradingpair':[],'r2':[],'rmse':[]})
    
   coinprice1 = symbol1 + '_price' 
   coin1_price = correlationData[[coinprice1]].head(1267)
   
   coinprice2 = symbol2 + '_price'
   coin2_price = correlationData[coinprice2].head(1267)
   tradingpair = symbol1 + '-' + symbol2

   X_train, X_test, y_train, y_test = train_test_split(coin1_price, coin2_price, test_size=0.30)

   #Training the model with set of predictors in X
   hypothesis = LinearRegression(fit_intercept=True) # hypothesis = SVR(kernel="rbf") # epsilon=0.0001)
   hypothesis.fit(X_train, y_train)

   # Prediction Scoring: r2_score and root mean squared error
   y_pred_train = hypothesis.predict(X_train)

   r2 = np.abs(r2_score(y_train, y_pred_train))
   rmse = np.sqrt(mean_squared_error(y_train, y_pred_train))

   df2 = df2.append({'tradingpair': tradingpair, 'r2': r2, 'rmse': rmse}, ignore_index=True)
   
   print(df2)

linearRegression('xrp','xlm')
```
